dobimo 4414 failure=1 instanc.. ostalih 4414 lahko izberemo random iz vseh datotek, da bo ucna mnozica dobro razporejena

```python
import pandas as pd
import glob
import os

vsiFili = glob.glob(os.path.join('data/**', "*.csv"), recursive=True)

seznam_odpovedi = []
skupno_failov = 0

for i, datoteka in enumerate(vsiFili):
    try:
        temp_df = pd.read_csv(datoteka, low_memory=False)

        odpovedi = temp_df[temp_df['failure'] == 1].copy()

        if not odpovedi.empty:
            seznam_odpovedi.append(odpovedi)
            skupno_failov += len(odpovedi)

        #da ne bo preve teksta, a e vedno ve, da se nekaj dogaja
        if i % 50 == 0:
            print(f"Obdelanih {i}/{len(vsiFili)} datotek... Trenutno odpovedi: {skupno_failov}")

    except Exception as e:
        print(f"Napaka pri branju {os.path.basename(datoteka)}: {e}")

if seznam_odpovedi:
    df_vsi_faili = pd.concat(seznam_odpovedi, ignore_index=True)
    print("\n" + "="*40)
    print(f"KONNI REZULTAT ZA CELO LETO:")
    print(f"Skupaj najdenih odpovedi (failure=1): {skupno_failov}")
    print("="*40)

    df_vsi_faili.to_csv('vseOdpovedi.csv', index=False)
    print("Vse odpovedi so shranjene v 'vseOdpovedi.csv'.")

    display(df_vsi_faili.head())
else:
    print("V nobeni mapi ni bilo najdenih odpovedi.")
```

ostala polovica ucne mnozice so diski ki so failure=0... izbiramo jih random

```python
import pandas as pd
import glob
import os
import random

vsiFili = glob.glob(os.path.join('data/', "**/*.csv"), recursive=True)
ciljnoStevilo = 4414

seznamZdravih = []
nabranihZdravih = 0
random.shuffle(vsiFili)

for datoteka in vsiFili:
    if nabranihZdravih >= ciljnoStevilo:
        break

    try:
        tempDf = pd.read_csv(datoteka, low_memory=False)
        zdravi = tempDf[tempDf['failure'] == 0]

        if not zdravi.empty:
            #iz vsake datoteke vzamemo najve 20 instanc
            vzorecVelikost = min(len(zdravi), 20)
            vzorec = zdravi.sample(n=vzorecVelikost)

            seznamZdravih.append(vzorec)
            nabranihZdravih += len(vzorec)

            if nabranihZdravih % 500 < 20: # Izpis napredka
                print(f"Nabranih: {nabranihZdravih}/{ciljnoStevilo} zdravih instanc...")

    except Exception as e:
        continue

dfZdravFinal = pd.concat(seznamZdravih, ignore_index=True).head(ciljnoStevilo)

dfModelReady = pd.concat([df_vsi_faili, dfZdravFinal], ignore_index=True)
#premesamo instance med sabo, da nebodo skupaj isti (failure=1/0..)
dfModelReady = dfModelReady.sample(frac=1).reset_index(drop=True)

print(f"Konni nabor podatkov (failure=1 + failure=0)...Skupaj vrstic: {len(dfModelReady)}")

# Shranimo konni nabor
dfModelReady.to_csv('koncniPodatkiZaModel.csv', index=False)
```
